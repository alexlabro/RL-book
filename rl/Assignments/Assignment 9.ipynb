```python
class OrderSimulator(MarkovProcess[bool]):
    '''A simple example Markov chain with two states, flipping from one to
    the other with probability p and staying at the same state with
    probability 1 - p.

    '''
    def __init__(self, Mid: float):
        self.Mid = Mid

    def transition(self, state: NonTerminal[OrderBook]) -> \
            Distribution[NonTerminal[OrderBook]]:
        def next_state(state=state):
            sell_buy = Bernoulli(0.5).sample()
            limit_market = Bernoulli(0.5).sample()
            shares = Uniform().sample()*40
            if limit_market :
                if sell_buy: 
                    price = Uniform().sample()*20 + self.Mid +1
                    ds, next_st = state.state.sell_limit_order(price, shares)
                else :
                    price = self.Mid - Uniform().sample()*20  - 1
                    ds, next_st = state.state.buy_limit_order(price, shares)
            else :
                if sell_buy: 
                    ds, next_st = state.state.sell_market_order(shares)
                else :
                    ds, next_st = state.state.buy_market_order(shares)
            return NonTerminal(next_st)

        return SampledDistribution(next_state)
```
